```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
categories = ['Token = 10', 'Token = 50', 'Token = 200']
metrics = {
    'CPUTime' : [9.8175, 11.33666667, 14.46],
    'Model' : [0.87278, 1.40868, 5.08992],
    'Block' : [0.8570075, 1.40109, 5.061],
    'MLP' : [0.555655, 0.8611266667, 3.07998],
    #'Attention' : [0.3104525, 0.4987166667, 1.9521]
}

attention = [0.3104525, 0.4987166667, 1.9521]
mlp = [0.555655, 0.8611266667, 3.07998]

# Plotting
plt.figure(figsize=(10, 6))
bottom = np.zeros(len(categories))

for metric, values in metrics.items():
    plt.bar(categories, values, label=metric)
    bottom += np.array(values)

print(bottom)
plt.bar(categories, attention, label='Attention', bottom=mlp)
    
plt.xlabel('Categories')

plt.ylabel('Latency (s)')
plt.title('Breakdown of GPT2 Program on Sapphire Rapids')
plt.legend()
plt.show()
```
